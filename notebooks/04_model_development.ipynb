```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Use the non-interactive backend for safety
import matplotlib
matplotlib.use('Agg')

print("AI Libraries Loaded!")
```

```python
# Load the data from processed folder
df = pd.read_csv('../data/processed/features_engineered.csv')

df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

print(f"Dataset ready with {df.shape[0]} days of history and {df.shape[1]} features.")
df.head()
```

```python
# Shifting the price back by 1 so 'today' has 'tomorrow's' price as a target
df['target'] = df['rate'].shift(-1)

data = df.dropna()

# Select the features we want the AI to look at, exclude the raw 'rate' to avoid 'data leakage'
features = ['lag_1', 'lag_7', 'ma_7', 'ma_30', 'volatility_7', 'day_of_week']
X = data[features]
y = data['target']

print("Features (X) and Target (y) are defined.")
```

```python
# Training yay
split_index = int(len(data) * 0.8)

X_train, X_test = X.iloc[:split_index], X.iloc[split_index:]
y_train, y_test = y.iloc[:split_index], y.iloc[split_index:]

print(f"Training Range: {X_train.index.min().date()} to {X_train.index.max().date()}")
print(f"Testing Range: {X_test.index.min().date()} to {X_test.index.max().date()}")
```

```python
from sklearn.ensemble import RandomForestRegressor

# n_estimators=100 means building 100 individual decision trees
model = RandomForestRegressor(n_estimators=100, random_state=42)

print("Training the AI (Random Forest)...")
model.fit(X_train, y_train)
print("Training Complete!")
```

```python
from sklearn.metrics import mean_absolute_error

# Make predictions
predictions = model.predict(X_test)

# Calculate MAE
mae = mean_absolute_error(y_test, predictions)
print(f"Average Error (MAE): {mae:.5f} GBP")

# Results and save image
plt.figure(figsize=(12, 6))
plt.plot(y_test.index, y_test.values, label='Actual', color='blue', alpha=0.5)
plt.plot(y_test.index, predictions, label='AI Prediction', color='red', linestyle='--', alpha=0.8)
plt.title('EUR/GBP Prediction vs Actual')
plt.legend()
plt.savefig('prediction_results.png')
print("Results saved to 'prediction_results.png'")
```
