```python
import pandas as pd
import numpy as np

# I import matplotlib.pyplot directly to see if the fix worked
try:
    import matplotlib.pyplot as plt
    import seaborn as sns
    # Use a safe backend that avoids the DLL issues 
    import matplotlib
    matplotlib.use('Agg')
    print("SUCCESS: Libraries loaded correctly!")
except AttributeError:
    print("ERROR: Matplotlib is still showing as broken. We may need to use a different version.")
except Exception as e:
    print(f"ANOTHER ERROR: {e}")

# Set visual style
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)
```

```python
# Load the dataset
df = pd.read_csv('../data/raw/eur_gbp_historical.csv')

# Convert date to datetime and set as index
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

# Print summary statistics
print("--- DATA OVERVIEW ---")
print(df.describe())
print("\n--- MISSING VALUES ---")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(15, 7))
plt.plot(df.index, df['rate'], label='EUR/GBP Rate', color='#1f77b4')
plt.title('EUR/GBP Historical Trend (2010-2026)')
plt.xlabel('Year')
plt.ylabel('Exchange Rate')
plt.grid(True, alpha=0.3)

# Save the plot 
plt.savefig('historical_trend.png')
print("Chart created and saved as 'historical_trend.png'")

# Display the plot in the notebook
plt.show()
```
